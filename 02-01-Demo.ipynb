### Conditional Probability


#### Q1. Given that  customer has taken term deposit, find the probability that she is a female.


```python
S = data.shape[0]
print('Total number of customers: ', S)
```

```python
pd.crosstab(data.gender, data.term_deposit, margins=True)
```

```python
# P(customer is a female | customer has taken a term deposit)
A = 20798
B = 4640
AnB = 2356
print('Number of female customers: P(A)', A)
print('Number of customer who have taken term deposit P(B):', B)
print('Customer is female and has taken term deposit: P(A&B)', AnB)

```

```python
pAnB = round(AnB/S,3)
pB = round(B/S, 3)
print('Joint probability that a customer is female has has taken term deposit:', pAnB )
print('Marginal probability that a customer has taken term deposit:', pB)
```

```python
# P(A | B) = P(A and B) / P(B)
pAgivenB = round(pAnB/pB, 3)
print('Conditional probability the customer is a female GIVEN customer has taken term deposit:',pAgivenB)
```
