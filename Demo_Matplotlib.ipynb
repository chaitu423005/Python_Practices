#Demo: **Data Visualization using Matplotlib**

###**Simple Plot with Matplotlib**

```python
import numpy, matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4]) #List of Co-ordinates
plt.show() #Displays plot
```

###**Simple Plot with Matplotlib using both axes**

```python

x =range(5) #Sequences of Values for the X-Axes
plt.plot(x, [x1**2 for x1 in x]) #Vertical Co-ordinates of the point plotted: y=x^2
plt.show()
```

```python

x = numpy.arange(0, 5, 0.01)
plt.plot(x, [x1**2 for x1 in x])
plt.show()
```

###**Multi-line Plots**

```python

x =range(5)
plt.plot(x, [x1 for x1 in x]) #Line 1
plt.plot(x, [x1*x1 for x1 in x]) #Line 2
plt.plot(x, [x1*x1*x1 for x1 in x]) #Line 3
plt.show()
```

```python

x =range(5)
plt.figure()
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])

plt.grid(True) #grid() function adds a grid to the plot. It takes a single Boolean Parameter
plt.show()
```

###**Some Important Attributes**

```python
#Limiting the Axes


x =range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 10)
plt.show()
```

```python
#Adding the Labels


x =range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlabel('X-axis') #X-label
plt.ylabel('Y-axis') #Y-label
plt.show()
```

```python
#Adding a Title

x =range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.show()
```

```python
# Adding a Legend

x = numpy.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend()
plt.show()
```

```python
#Saving the Plots

x = numpy.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend()
plt.savefig('plot.png') #savefig() function saves the plot for further use
plt.show()
```

```python
#Control Colours

import numpy as np

y = np.arange(1, 3)
plt.plot(y, 'b') # Blue Color
plt.plot(y+5, 'k') # Black Color
plt.plot(y+10, 'y') #Yellow Color
plt.show()
```

```python
# Control Line Styling

import numpy as np
y = np.arange(1, 100)
plt.plot(y, '--', 
         y*5,'-.', 
         y*10, ':')
plt.show()
```

```python
# Control Marker Styling

y = np.arange(1, 3, 0.2)
plt.plot(y, '*', 
         y+0.5, 'o',
         y+1, 'D', 
         y+2, '^', 
         y+3, 's')
plt.show()
```

###**Types of Plots**

####**Histogram** 
Displays the distribution of a variable over a range of frequencies or values

```python
import matplotlib.pyplot as plt, numpy
y = numpy.random.randn(100, 100) #100x100 array of a Gaussian Distribution
plt.hist(y) #Function to plot the Histogram
plt.show()
```

####**Bar Chart**
Visually compare two or more values using Rectangular Bars

```python
import matplotlib.pyplot as plt
plt.bar([1, 2, 3], [1, 4, 9])
plt.show()
```

```python
#Plotting a Dictionary Using Bar Plot/Chart

import matplotlib.pyplot as plt, numpy
dictionary = {'A':25,'B':70,'C':55,'D': 90}
for i, key in enumerate(dictionary):
    plt.bar(i, dictionary[key])
    plt.xticks(numpy.arange(len(dictionary)), 
               dictionary.keys())
plt.show()
```

####**Scatter Plot**
Display values for two sets of data, visualized as a collection of points

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.scatter(x, y)
plt.show()
```

####**Heat Map**
Used to visualize patterns of events in a dataset

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.random.random((16, 16))

#HeatMaps are plotted using the imshow() function in Matplotlib

plt.imshow(a, cmap='hot', interpolation='nearest') 
plt.show()
```

####**Pie Chart**
Used to compare proportionality in data or visualize categorical data



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
x = [40, 20, 5]
labels = ['Bikes', 'Cars', 'Buses']
plt.pie(x, labels=labels)
plt.show()
```
