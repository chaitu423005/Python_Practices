###**Dataset Description**

The FBI and the Centers for Disease Control and Prevention log fatal shootings by police, but officials acknowledge that their data is incomplete. In 2015, The Post documented more than two times more fatal shootings by police than had been recorded by the FBI. Last year, the FBI announced plans to overhaul how it tracks fatal police encounters

The dataset contains the following attributes:

- **id** - Unique ID 
- **name** - Name of the Victim
- **date** - Date of the Shootout
- **manner of death** - Manner of Death i.e., Shot or Shot & Tasered
- **armed** - Arm
- **age** - Age of the Victim
- **gender** - Gender
- **race** - Race of the Victim
       W: White, non-Hispanic 
       B: Black, non-Hispanic 
       A: Asian 
       N: Native American 
       H: Hispanic 
       O: Other 
       
- **city** - City in which the Shootout happened
- **state** - State in which the Shootout happened

```
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")


df = pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

###**Question 2:**
Find out the **Age Span**, **Gender**, and **Race** which has the highest number of deaths in the shootout using a Bar Chart

```
plt.figure(figsize=(12,8))
plt.hist(df.age)
plt.title('Most Victims By Age')
plt.tight_layout()
plt.show()
```

The data shows the age of people being killed is between 20-40

```
df_new = df.groupby(['gender'])['age'].count()

plt.figure(figsize=(12,8))
df_new.plot.bar(color=['pink', 'blue'], width = 0.2)
plt.title('Most Victims By Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

From above, we can infer that most of the Victims are **Male**

```
df_new = df.groupby(['race'])['manner_of_death'].count()

plt.figure(figsize=(12,8))
df_new.plot.bar(color=['black', 'red', 'green', 'blue', 'cyan','yellow'])
plt.title('Most Victims By Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

From above, we can infer that most of victims are **White** followed by Black, Hispanic and others

###**Question 3:**
Find the **Top 10 States** with the most number of shootings using a **Bar Plot** and a **Pie Chart**

```
# Top 10 states with most shootings
topten_state = df['state'].value_counts()[:10]
topten_state = pd.DataFrame(topten_state).reset_index()

plt.figure(figsize=(12,8))
plt.bar(topten_state['index'], topten_state.state,color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xlabel('States')
plt.ylabel('No. of Deaths')
plt.title('Top 10 States with the Most Deaths')

plt.tight_layout()

plt.show()
```

From above, we can infer that California is the state with the most number of shootings

```
plt.figure(figsize=(12,8))

plt.pie(topten_state.state, labels=topten_state['index'],autopct='%1.1f%%')
plt.tight_layout()
plt.show()
```

From above, we can infer that **California** has most deaths i.e., 27 % of the total shootings followed by Texas and Florida

Let's explore the State of California a bit more

###**Question 4:**
Find out how many Police Officers was wearing **body_camera** in the **State of California** during the shootout 

```
state_grp  = df.groupby(['state'])
new = state_grp['body_camera'].value_counts(normalize=True).loc['CA']

plt.figure(figsize=(12,8))


new.plot.bar(color = ['red', 'yellow'], width = 0.2)
plt.title('Body Camera Status in the State of California')
plt.ylabel('No. of Deaths')

plt.tight_layout()

plt.show()
```

From above, we can infer that, in more than **80 %** of the cases, the Police Officer was not wearing **body_camera** during the shootout in the State of California which is very suspicious

###**Question 5:**
Find out which particular race has the highest number of deaths in the **State of California** in Police Shootouts

```

new = state_grp['race'].value_counts().loc['CA']

plt.figure(figsize=(12,8))
new.plot.bar(color = ['red', 'green', 'yellow', 'cyan'])
plt.title('Victims by Race in the State of California')
plt.xlabel('Race')
plt.ylabel('No. of Deaths')
plt.tight_layout()

plt.show()
```

From above, we can infer that in the state of California, most of the people killed in Police shootouts are **Hispanic (H)** followed by **White (W)** and **Black (B)** and so on

###**Question 6:**
Find out the Top 5 Cities with the most shootings or the number of deaths

```
# Top 5 Cities with most shootings
topfive_city = df['city'].value_counts()[:5]
topfive_city = pd.DataFrame(topfive_city).reset_index()

plt.figure(figsize=(12,8))
plt.bar(topfive_city['index'], topfive_city.city,color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xlabel('City')
plt.ylabel('No. of Deaths')
plt.title('Top Five Cities with the Most Deaths')
plt.tight_layout()

plt.show()
```

From above, we can infer that southern state cities such as Los Angeles have most no. of deaths 

Florida, as a state is number three with most deaths but we do not see any city from Florida on top five. It appear death numbers are divide across all cities in Florida

###**Question 7:**
Find out what percentage of Police Officers were wearing **body_camera** during the shootout 

```

df_new = df['body_camera'].value_counts(normalize = True)


plt.figure(figsize=(12,8))
df_new.plot.bar(color=['red', 'yellow'])
plt.title('Body Camera Status')
plt.xlabel('Body Camera Status')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

From above, we can infer that in more than **80 %** of the shootings, the police were not wearing body_camera

###**Question 8:**
Find out how many victims tried to flee 

```
df_new = df['flee'].value_counts()

plt.figure(figsize=(12,8))
df_new.plot.bar(color=['red', 'yellow','blue','cyan'])
plt.title('Fleeing Status')
plt.xlabel('Fleeing Status')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

From above, we can infer that the cases of **Not fleeing** are **HIGH** whereas the cases of fleeing are at a low number

###**Question 9:**

Visualize the Threat Level posed by the Victims

```
df_new = df['threat_level'].value_counts(normalize = True)

plt.figure(figsize=(12,8))
df_new.plot.bar(color=['red', 'yellow','blue','cyan'])
plt.title('Threat Level')
plt.xlabel('Threat Level')
plt.ylabel('Count')
plt.tight_layout()

plt.show()
```

From above, we can infer the various threat level posed by the people killed in the shootout

###**Question 10:**

Find out the weapons most used by the Victims during the shootout   

```
df_new = df['armed'].value_counts()[:5]
df_new = pd.DataFrame(df_new).reset_index()


plt.figure(figsize=(12,8))
plt.pie(df_new.armed, labels=df_new['index'],autopct='%1.1f%%')
plt.tight_layout()
plt.show()
```

From above, we can infer that in most of the shootings the victim had a **gun** or a **knife((. But, it is also noticeable that many were **unarmed** or had a **toy weapon** 

###**Conclusion:**
After analyzing and visualizing the dataset, we can conclude that

- The Police Officers were not wearing **body_camera** in most of the cases which is very suspicious
- In the State of California, the number of cases involving people of **Hispanic** race is unusually high as compared to other races
- In a significant amount of cases, the victim was either **Unarmed** or had a **toy gun**
